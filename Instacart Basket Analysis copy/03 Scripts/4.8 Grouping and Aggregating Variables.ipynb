#### Question 4.8.4

```python
## Apply transform function

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'

## Checking the data frame

ords_prods_merge.head(10)
```

```python
## Using loyalty flag to find the avearge prices of the purchased products

ords_prods_merge.groupby('loyalty_flag')['prices'].mean()
```

##### It appears that the loyal customers have purchased the products multiple times. However, the order value of the products are lesser compared to the new and regular customers

#### Question 4.8.6

```python
## Creating spending flag for each user based on the average price

ords_prods_merge['spending_behavior'] = ords_prods_merge.groupby('user_id')['prices'].transform('mean')
```

```python
## Creating the spending flag using .loc function

ords_prods_merge.loc[ords_prods_merge['spending_behavior'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['spending_behavior'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.head()
```

#### Question 4.8.7

```python
## Creating order frequency flag

ords_prods_merge['order_frequent'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequent'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['order_frequent'] > 10) & (ords_prods_merge['order_frequent'] <= 20), 'order_freq_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['order_frequent'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```
