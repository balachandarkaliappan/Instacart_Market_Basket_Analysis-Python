# Task - Exercise 4.6. Combining & Exporting Data

```python
## Import the packages

import pandas as pd
import numpy as np
import os
```

#### Task 4.6.3: Import the data file

```python
## Import the data files
path = r'/Users/balachandark/Desktop/Instacart Basket Analysis/02 Data'
orders_products_combined = pd.read_pickle(os.path.join(path, 'Prepared Data', 'orders_products_combined.pkl')) 
prods_df = pd.read_csv(os.path.join(path, 'Prepared Data', 'products_checked.csv'), index_col = False)
```

#### Task 4.6.4: Check the shape of the imported data frame

```python
orders_products_combined = orders_products_combined.drop('Unnamed: 0', axis = 1)
orders_products_combined.shape
```

```python
orders_products_combined.head()
```

```python
prods_df.shape
```

```python
prods_df = prods_df.drop('Unnamed: 0', axis = 1)
prods_df.head()
```

```python
orders_products_combined.head()
```

#### Task 4.6.4 & 4.6.5: Merging data files

```python
ords_prods_merge = orders_products_combined.merge(prods_df , on = ['product_id'] , indicator = 'match' )
```

```python
# Drop the _merge column

ords_prods_merge.drop('_merge', axis =1)
```

#### Task 4.6.6: Check the merge flag

```python
ords_prods_merge['_merge'].value_counts()
```

#### Task 4.6.7: Export the file

```python
ords_prods_merge.to_pickle(os.path.join(path, 'Prepared Data', 'ords_prods_merged.pkl'))
```
